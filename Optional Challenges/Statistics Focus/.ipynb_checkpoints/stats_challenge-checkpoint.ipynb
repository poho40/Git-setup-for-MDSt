```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read in data
df = pd.read_csv("./orders.csv")
print(df)
```

```python
#filter to last 30 days
last30Days = df["order_value"].mean()
f = plt.figure(figsize=(12,6))
#There are multiple items in an order and some orders are in the hundreds of thousands ad can be seen below,
#while the majority are less than 1000.  This skews the data to the left which increases the average order value
plt.scatter(df["total_items"], df["order_value"])
plt.xlabel('total_items')
plt.ylabel('order_value')
```

```python
# Find the median order value instead as the median is more resistant to outliers
```

```python
print((df["order_value"]).median())
```

```python
import matplotlib.dates as mdates

#shows all points
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
plt.scatter((df["created_at"].astype('datetime64')), df["order_value"])
plt.xlabel('created_at')
plt.ylabel('order_value')
```

```python
import matplotlib.dates as mdates

#shows only orders less than $1000
plt.ylim(0,1000)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
plt.scatter((df["created_at"].astype('datetime64')), df["order_value"])
plt.xlabel('created_at')
plt.ylabel('order_value')
```

```python
The customer trend is relatively consistent with most people ordering items worth less than $1000 total, with some people
ordering a lot every couple of days.
```
